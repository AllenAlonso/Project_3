```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup as soup
import matplotlib.pyplot as plt
import pandas as pd
from webdriver_manager.chrome import ChromeDriverManager
import time
```

```python
browser = Browser('chrome',executable_path=ChromeDriverManager().install())
```

```python
# Visit the website
url = "https://www.thebaseballcube.com/page.asp?PT=draft_round&ID=2022~1~June-Reg"
browser.visit(url)
```

```python
# Create a Beautiful Soup Object
html = browser.html
draft_picks = soup(html, 'html.parser')
```

```python
df_list = []
#loop through the year data s
for y in range(1,6):
        #selecting/clicking correct year 
        year = browser.find_by_id('dd4').first
        year.find_by_tag('option')[y].click()
        #selecting/clicking first round of draft of selected year
        round = browser.find_by_id('dd3').first
        round.find_by_tag('option')[1].click()
        
        time.sleep(4)

        # Create a Beautiful Soup Object      
        html = browser.html
        draft_picks = soup(html, 'html.parser')
        browser.find_by_id('csv_text_dg_0').click()
        
        #Getting all table data
        table = draft_picks.find('table', id = 'dg_0')
        rows = table.find_all('tr')
        table_headers = [row.text.strip() for row in rows[0].find_all('td')]  
        table_data = [[data.text.strip() for data in row.find_all('td')] for row in rows[1:]]
        #Adding DataFrame into the df list
        df_list.append(pd.DataFrame(table_data, columns=table_headers))
        
        #Check print
        print(f'--- dataframe {y} ---')

        #Looping through rounds 2-10
        for r in range(2, 11):
                #Selecting/Clicking round r 
                round = browser.find_by_id('dd3').first
                round.find_by_tag('option')[r].click()

                time.sleep(4)
                
                # Create a Beautiful Soup Object
                html = browser.html
                draft_picks = soup(html, 'html.parser')
                browser.find_by_id('csv_text_dg_0').click()

                #Getting all table data
                table = draft_picks.find('table', id = 'dg_0')
                rows = table.find_all('tr')
                table_headers = [row.text.strip() for row in rows[0].find_all('td')]  
                table_data = [[data.text.strip() for data in row.find_all('td')] for row in rows[1:]]
                temp = pd.DataFrame(table_data, columns=table_headers)
                #Adding DataFrame into the df list from the same year
                df_list[y-1] = pd.concat([df_list[y-1],temp])
        
```
